```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fs = 1
N = 100
t = np.arange(N)
s = np.sin(0.15*2*np.pi*t)
plt.plot(t, s)
```

```python
s = s * np.hamming(100)
S = np.fft.fft(s, 1024)
f = np.arange(-Fs/2, Fs/2, Fs/1024)

S_shift = np.fft.fftshift(S)
plt.figure()
plt.title("MAGA")
S_shift_mag = np.abs(S_shift)
plt.plot(f,S_shift_mag,'.-')
# plt.ylim([-0.01,0.005])
# plt.xlim([0.9, 0.19])

plt.figure()
plt.title("phase")
S_shift_phase = np.angle(S_shift)
plt.plot(f,S_shift_phase,'.-')
plt.show()

```
