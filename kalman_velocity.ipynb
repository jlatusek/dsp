```python
import numpy as np
from pypdf._text_extraction import orient
from scipy import signal
import matplotlib.pyplot as plt
```

```python
f1 = 2000
Fs = 48000
N = 2**8
t = np.linspace(0, N/Fs, N)

orig_sig = np.sin(f1*2*np.pi*t)
plt.figure()
plt.title("Original signal")
plt.plot(t, orig_sig)
plt.show()
```

```python
noise_amplitude = 0.1
noise = np.random.normal(0, 1, orig_sig.shape) * noise_amplitude
signal = orig_sig + noise

plt.figure()
plt.title("Noisy signal")
plt.plot(t, signal)
plt.show()

```
