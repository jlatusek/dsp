```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
sig = np.concatenate((np.zeros([20, 1]), np.ones([20,1]), np.zeros([20,1])), axis=None)

plt.figure()
plt.plot(sig, '.-')
```

```python
sig_len = len(sig)
convolution = np.zeros([sig_len*2, 1]) 
sig2 = None

for i in range(1, sig_len):
    sig2 = np.concatenate((sig[-i:], np.zeros([sig_len-i, 1])), axis=None)
    convolution[i] = np.sum(sig * sig2)

j = 60
for i in range(sig_len, 1, -1):
    sig2 = np.concatenate( (np.zeros([sig_len - i, 1]), sig[:i]), axis=None)
    convolution[j] = np.sum(sig * sig2)
    j +=1

plt.figure()
plt.plot(convolution)
plt.show()
```
